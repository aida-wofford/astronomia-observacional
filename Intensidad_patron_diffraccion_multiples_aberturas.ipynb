### Ejemplo de cómo graficar la función sen(theta)

```python
fig = plt.figure(figsize=(5,3)) #Definimos el tamaño de la figura
y = np.sin(theta)
plt.plot(theta, y, marker='.')
plt.xlabel(r'$\theta$')
plt.ylabel(r'Sen ( $\theta$ )')
plt.show()
```

### Definimos los parámetros de la ecuación 4.1 y creamos un arreglo de ángulos theta

```python
D = 1 #tamaño abertura
d = 20 #distancia entre aberturas, número de picos
N = 5 #numero de aberturas
theta = np.arange(-0.05*np.pi, 0.05*np.pi, 0.001)# distancia angular respecto al eje óptico
theta = np.arange(-0.05*np.pi, 0.05*np.pi, 0.001)# distancia angular respecto al eje óptico
```

### Graficamos el resultado para dos longitudes de onda distintas

```python
fig = plt.figure(figsize=(15,3))

for i in range(2):
# Escogemos la longitud de onda cuyo patrón de difracción vamos a graficar
    if i == 0: lam = 1.0 #longitud de onda más azul: segunda franja más cerca del eje óptico  
    if i == 1: lam = 1.1 #longitud de onda más roja: segunda franja más lejos del eje óptico
    a = D*np.pi*np.sin(theta)/lam
    b = N*d*np.pi*np.sin(theta)/lam
    c = d*np.pi*np.sin(theta)/lam
    y = (((np.sin(a))**2) / (a**2)) * (((np.sin(b))**2) / ((np.sin(c))**2))
    plt.plot(theta, y, label='$\lambda$='+str(lam))
    plt.xlabel(r'$\theta$')
    plt.ylabel(r'I ( $\theta$ )')

plt.legend()
plt.show()
```

### Solución al ejercicio b1

```python
D = 1 #tamaño abertura
d = 20 #distancia entre aberturas, número de picos
lam = 1. #longitud de onda
theta = np.arange(-0.05*np.pi, 0.05*np.pi, 0.001)# distancia angular del eje óptico


fig = plt.figure(figsize=(15,3))

for i in range(2):
# Escogemos la longitud de onda cuyo patrón de difracción vamos a graficar
    if i == 0: N = 2 #numero de aberturas menor, menor intensidad de la líneas  
    if i == 1: N = 3 #numero de aberturas mayor, mayor intensidad de la líneas
    a = D*np.pi*np.sin(theta)/lam
    b = N*d*np.pi*np.sin(theta)/lam
    c = d*np.pi*np.sin(theta)/lam
    y = (((np.sin(a))**2) / (a**2)) * (((np.sin(b))**2) / ((np.sin(c))**2))
    plt.plot(theta/(lam/(2*d)),y, label='N='+str(N))
    plt.xlim(-0.9, 6.9)
    plt.xlabel(r'$\theta$ en unidades de $\lambda/(2d)$')
    plt.ylabel(r'I ( $\theta$ )')

plt.legend()
plt.show()
```

### Solución al ejercicio c1

```python
D = 1 #tamaño abertura
d = 20 #distancia entre aberturas, número de picos
lam = 1. #longitud de onda
theta = np.arange(-0.05*np.pi, 0.05*np.pi, 0.001)# distancia angular del eje óptico


fig = plt.figure(figsize=(15,3))

for i in range(2):
# Escogemos la longitud de onda cuyo patrón de difracción vamos a graficar
    if i == 0: D = 5 #menor distancia, más intensidad en los máximos 
    if i == 1: D = 50 #mayor distancia, menos intensidad en los máximos
    a = D*np.pi*np.sin(theta)/lam
    b = N*d*np.pi*np.sin(theta)/lam
    c = d*np.pi*np.sin(theta)/lam
    y = (((np.sin(a))**2) / (a**2)) * (((np.sin(b))**2) / ((np.sin(c))**2))
    plt.plot(theta/(lam/(2*d)),y, label='D='+str(D))
    plt.xlim(-0.9, 6.9)
    plt.xlabel(r'$\theta$ en unidades de $\lambda/(2d)$')
    plt.ylabel(r'I ( $\theta$ )')

plt.legend()
plt.show()
```
