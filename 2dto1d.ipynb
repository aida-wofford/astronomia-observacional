As an example we will use a 2d spectrum of star-forming region NGC 346 that was obtained at the VLT with instrument FORS1.

```python
spectrum2d_filename = 'https://github.com/aida-wofford/astronomia-observacional/blob/main/32030000.346C.16.3.R.W.fits?raw=true'
```

# Import and visualize FITS image

The image is stored as a binary FITS file. A FITS file is composed of several units that contain different information. Each unit has a header that gives information about the content of the unit. Examples of content are the science data, the data quality flags, etc.. Our first task is to learn to import, parse, and visualize the images. For this we will use the astropy (<a href="https://docs.astropy.org/en/stable/io/fits/">AstroPy Documentation</a>) and numpy (<a href="https://numpy.org/">NumPy Documentation</a>) packages.

```python
import astropy.io
import numpy as np
from astropy.io import fits
# Header Data Unit List
hdul = fits.open(spectrum2d_filename)
```

<code>HDUList.info()</code> summarizes the contents of the opened FITS file. We can use this to check that we have successfully loaded a picture and find out its format and dimensions.

```python
# summarize the content of the opened FITS file
hdul.info()
```

The image data is a numpy array found in the data property of the first Header Data Unit.


```python
image_data = hdul[0].data
```

Now that we have imported the data, we can close the file.


```python
hdul.close()
```

We can visualize the data using matplotlib pyplot.imshow() (<a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html">imshow</a>).


```python
from matplotlib import pyplot as plt
%matplotlib inline 
plt.imshow(image_data, interpolation='nearest')
plt.colorbar()
plt.title('2d spectrum', fontsize=10)
plt.show()
```

We want to change to grey scale and only see pixels that have counts either 0 or less, 1, 2, ..., or 100 and more. 

```python
plt.imshow(image_data,vmin=0,vmax=100,cmap='gray')
plt.colorbar()
plt.show()
```

What happens if we set the scale from vmin=0 to vmax=1000? 

```python
plt.imshow(image_data,vmin=0,vmax=1000,cmap='gray')
plt.colorbar()
plt.show()
```

We are seeing the pixels that are closer to 1000 because the pixels with 10 counts are dimmer.
